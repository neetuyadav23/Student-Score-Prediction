# Student Score Prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
print(os.getcwd())  # Shows the folder where Python is looking for the file
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())  # Display first few rows
```

```python
plt.scatter(df['Hours Studied'], df['Scores'], color='red')
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.title("Study Hours vs Scores")
plt.show()
```

```python
X = df['Hours Studied'].values  # Extract as NumPy array
y = df['Scores'].values
```

```python
N = len(X)

# Calculate slope (m) and intercept (b)
m = (N * np.sum(X * y) - np.sum(X) * np.sum(y)) / (N * np.sum(X**2) - (np.sum(X))**2)
b = (np.sum(y) - m * np.sum(X)) / N

print(f"Calculated Slope (m): {m}")
print(f"Calculated Intercept (b): {b}")
```

```python
y_pred = m * X + b
```

```python
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.title("Linear Regression: Study Hours vs Scores")
plt.legend()
plt.show()
```

```python
mse = np.mean((y - y_pred) ** 2)
print(f"Mean Squared Error: {mse}")
```

```python
ss_total = np.sum((y - np.mean(y))**2)
ss_residual = np.sum((y - y_pred)**2)
r2_score = 1 - (ss_residual / ss_total)

print(f"R² Score: {r2_score}")
```

```python
new_hours = 6.5
predicted_score = m * new_hours + b
print(f"Predicted Score for {new_hours} hours of study: {predicted_score}")
```
